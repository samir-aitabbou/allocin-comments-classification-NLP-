# Cleaning

```python
import re
import nltk
from nltk.corpus import stopwords

# Charger la liste de stopwords en français
nltk.download('stopwords')
stop_words = set(stopwords.words('french'))

def nettoyer_texte(texte):
    # Conversion en minuscules
    texte = texte.lower()
    
    # Suppression des caractères spéciaux, des chiffres, des liens URL, des mentions et des hashtags
    
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    texte = url_pattern.sub('', texte)
    
    texte = re.sub(r'[^a-zA-Z]', ' ', texte)
    texte = re.sub(r'@\w+', '', texte)
    texte = re.sub(r'#\w+', '', texte)
    
    # Suppression des stopwords
    mots = texte.split()
    mots_filtrés = [mot for mot in mots if mot not in stop_words]
    
    return ' '.join(mots_filtrés)

```

```python
# Concaténer les données d'apprentissage et de test pour former le corpus entier
corpus = pd.concat([train_data, dev_data,test_data],axis=0,ignore_index=True)
```

```python
corpus 
```

```python
corpus['commentaire'] = corpus['commentaire'].astype(str)
```

```python
corpus['commentaire'] = corpus['commentaire'].apply(nettoyer_texte)
```

```python
corpus
```

```python
# Obtenir le lexique (liste de tous les mots uniques)
lexique = set(' '.join(corpus['commentaire']).split())
```

```python
len(lexique)
```

```python
lexique
```

```python
# Créer un dictionnaire pour mapper chaque mot du lexique à un numéro unique
mot_vers_numero = {mot: i + 1 for i, mot in enumerate(lexique)}
```

```python
mot_vers_numero
```

```python
# Fonction pour décompter les occurrences de chaque mot dans un message

def compter_occurrences(message):

    occurrences_dict = {}  # Dictionnaire pour stocker l'indice et le nombre d'occurrences de chaque mot

    
    mots = message.split()
    
    for mot in mots:
        if mot in mot_vers_numero:
            index = mot_vers_numero[mot] - 1
            if index in occurrences_dict:
                occurrences_dict[index] += 1
            else:
                occurrences_dict[index] = 1
    
    return occurrences_dict

    
# # Appliquer la fonction pour chaque message dans le corpus entier
corpus['Occurrences'] = corpus['commentaire'].apply(compter_occurrences)
```
